```python
import pandas as pd
```

```python
df = pd.read_csv('Result.csv')
df
```

```python
def preprocess(contentArray):
  preProcRes = []
  for content in contentArray:
    new = str(content).replace("\n", " ")
    preProcRes.append(new)
  return preProcRes
```

```python
preProcArray = preprocess(df["Content"].values)
```

```python
from summarizer import Summarizer,TransformerSummarizer

results = []

GPT2_model = TransformerSummarizer(transformer_type="GPT2",transformer_model_key="gpt2-medium")

for content in preProcArray:
  results.append(''.join(GPT2_model(content, min_length=60)))

print(results)
```

```python
df["Summarized"] = results
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
df.to_csv("summ.csv",index=False)
```
